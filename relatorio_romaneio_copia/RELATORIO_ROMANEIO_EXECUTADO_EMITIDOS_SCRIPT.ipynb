```python
import pandas as pd 
def romaneios_executados(caminho1, caminho2):
    #define os dataframes baseado nos dados fornecidos
    romaneio_executado = pd.read_excel(caminho1)
    romaneio_emitido = pd.read_excel(caminho2)
    #exclui a primeira linha do dataframe
    romaneio_emitido.drop([0], axis= 0 , inplace=True)
    #cria um novo dataframe baseado na coluna, romaneio do 1 dataframe, FRT.AGREGADO APURADO do segundo dataframe
    romaneio_final = pd.merge(romaneio_executado[["ROMANEIO"]], 
                          romaneio_emitido[["ROMANEIO", "FRT.AGREGADO APURADO"]],
                          on="ROMANEIO", 
                          how="left")
    #renomeia a coluna para apurado
    romaneio_final.rename({"FRT.AGREGADO APURADO":"APURADO"}, axis=1, inplace=True)
    
    #cria um novo dataframe e utiliza a função merge, para juntar os dataframes com base na coluna romaneio
    romaneio_executado = pd.merge(romaneio_executado, romaneio_final, on="ROMANEIO", how="left")
    
    #muda o tipo de dado da coluna romaneio para string e adiciona um campo de 6 valores padrões para os valores nas celulas
    romaneio_executado['ROMANEIO'] = romaneio_executado['ROMANEIO'].astype(str)
    romaneio_executado['ROMANEIO'] = romaneio_executado['ROMANEIO'].str.zfill(6)
    
    #salva o arquivo
    romaneio_executado.to_excel("Romaneios_Março_BHZ_teste.xlsx")
    #faz o retorno do dataframe finalizado 
    return romaneio_executado
    
#variaveis utilizadas para especificar os caminhos a função   
caminho1 = "ROMANEIO_EXECUTADOS_SPO.XLSX"
caminho2 = "ROMANEIO_EMITIDOS_SPO.XLSX"
#executando a função e joga o resultado(dataframe) em uma nova variavel
romaneio_executado = romaneios_executados(caminho1, caminho2)
#exibe o dataframe finalizado
display(romaneio_executado)

```
