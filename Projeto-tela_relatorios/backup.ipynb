```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox

# Função para procurar arquivos
def procurar_arquivos():
    file_names = filedialog.askopenfilenames()

# Função para chamar a tela principal
def mostrar_tela_principal():
    tela_secundaria.pack_forget()
    tela_principal.pack(fill="both", expand=True)

# Função para chamar a tela secundária
def mostrar_tela_secundaria():
    tela_principal.pack_forget()
    tela_secundaria.pack(fill="both", expand=True)

# Criando a tela principal
root = tk.Tk()
root.title("Sistema de Relatórios")
root.geometry("1024x768")
root.configure(bg="DarkGray")
root.resizable(False,False)

# Criando tela principal
tela_principal = tk.Frame(root, bg="DarkGray")

# Criando a label da tela principal
label_informativa = tk.Label(tela_principal, text="Selecione a opção desejada", bg="DarkGray", font=("League Spartan", 15, "bold", "italic"))
label_informativa.pack(pady=150)

# Botão iniciar a dela principal
botao_iniciar = tk.Button(tela_principal, text="Iniciar", width=30, height=2, command=mostrar_tela_secundaria)
botao_iniciar.pack(pady=200)

# Criando tela secundária
tela_secundaria = tk.Frame(root, bg="DarkGray")

# Label da tela secundária
label_secundaria = tk.Label(tela_secundaria, text="Segunda Tela", bg="DarkGray", font=("League Spartan", 15, "bold", "italic"))
label_secundaria.pack()

# Botão para iniciar a busca de arquivos
browse_button = tk.Button(tela_secundaria, text="Procurar Arquivos", command=procurar_arquivos, width=25, height=2)
browse_button.place(x=300, y=600)

# Botão para voltar a tela principal
back_button = tk.Button(tela_secundaria, text="Voltar ao Menu Inicial", command=mostrar_tela_principal, width=25, height=2)
back_button.place(x=100, y=600)

# Botão para iniciar os relatório de apuração de lucratividade e de romaneios executados
botao_apuracao_lucra = tk.Button(tela_secundaria, text="Apuracao Lucratividade", width=25, height=2)
botao_apuracao_lucra.place(x=100, y=550)

botao_romaneio_execu = tk.Button(tela_secundaria, text="Romaneios Executados", width=25, height=2)
botao_romaneio_execu.place(x=300, y=550)

# Mostrar tela principal inicialmente
mostrar_tela_principal()

# Mantendo a tela executada
root.mainloop()
```
