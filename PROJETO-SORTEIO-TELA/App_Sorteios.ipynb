```python
import tkinter as tk
from tkinter import messagebox
import random

# Lista para armazenar os resultados registrados
registros_resultados = []

# Função para sortear e exibir resultados
def sortear():
    # Pegar os valores das entradas
    numeros_str = entrada_numeros.get()
    strings_str = entrada_strings.get()

    # Verificar se há entrada de números
    if numeros_str:
        lista_numeros = numeros_str.split(',')
        # Embaralhar a lista de números
        random.shuffle(lista_numeros)
    else:
        messagebox.showerror("Erro", "Por favor, insira números")
        return

    # Verificar se há entrada de strings
    if strings_str:
        lista_strings = strings_str.split(',')
        # Embaralhar a lista de strings
        random.shuffle(lista_strings)
    else:
        lista_strings = []

    # Combina as duas listas em uma lista de tuplas
    resultados_dict = dict(zip(lista_numeros, lista_strings))
    
    # Atualizar os resultados registrados
    registros_resultados.extend(resultados_dict.items())

    # Exibir os resultados com uma contagem regressiva entre cada exibição
    exibir_resultado_com_contagem(resultados_dict)
        
# Função para exibir resultados com contagem regressiva
def exibir_resultado_com_contagem(resultados_dict):
    if resultados_dict:
        numero, string = resultados_dict.popitem()  # Pega o último item do dicionário

        # Atualizar o texto do rótulo com o resultado
        resultados_label.config(text=f"Número: {numero}, Time: {string}")

        # Atualizar a interface gráfica
        tela_principal.update()

        # Aguardar um tempo antes de exibir o próximo resultado (em milissegundos)
        tela_principal.after(2000, exibir_resultado_com_contagem, resultados_dict)  # 3000 milissegundos = 3 segundos
    else:
        resultados_label.config(text="Sorteio concluído!")

# Função para exibir resultados registrados na janela auxiliar
def exibir_resultados_registrados():
    # Pegar os resultados do registro e formatá-los
    resultados_formatados = ""
    for numero, string in registros_resultados:
        resultados_formatados += f"NÚMERO: {numero}, TIME: {string}\n"
    
    # Atualizar o rótulo com os resultados registrados
    resultados_registrados_label.config(text=resultados_formatados)
    janela_resultados.deiconify()  # Exibir a janela de resultados

# Tela principal
tela_principal = tk.Tk()
tela_principal.title("SORTEADOR")
largura = 800
altura = 500
tela_principal.geometry(f"{largura}x{altura}")
tela_principal.resizable(False,False)
tela_principal.configure(bg="LemonChiffon")

# Criar rótulo para exibir resultados
resultados_label = tk.Label(tela_principal, text="", font=("Helvetica", 12, "bold"), bg="LemonChiffon", wraplength=700, justify="left")
resultados_label.place(x=50, y=400)

# Janela auxiliar para mostrar os registros de dados
janela_resultados = tk.Toplevel(tela_principal)
janela_resultados.title("Resultados Registrados")
janela_resultados.geometry("400x300")
janela_resultados.withdraw()  # Ocultar a janela inicialmente

# Chamar a função para receber os dados em outra janela
resultados_registrados_label = tk.Label(janela_resultados, text="", font=("Helvetica", 12, "bold"), wraplength=380, justify="left")
resultados_registrados_label.pack(pady=20)

# Criar e posicionar os objetos
tk.Label(tela_principal, text="SORTEIOS", font=("Arial Black", 15, "bold"), bg="LemonChiffon").place(x=350,y=20)
tk.Label(tela_principal, text="DIGITE OS NUMEROS (Separado por virgulas):", font=("Helvetica", 12, "bold"), bg="LemonChiffon").place(x=44,y=145)
entrada_numeros = tk.Entry(tela_principal, width=59)
entrada_numeros.place(x=405, y=150)

# Criar e posicionar os objetos
tk.Label(tela_principal, text="DIGITE OS NOMES (Separado por vírgula (opcional)):", font=("Helvetica", 12, "bold"), bg="LemonChiffon").place(x=44,y=195)
entrada_strings = tk.Entry(tela_principal, width=51)
entrada_strings.place(x=453,y=200)

# Botão para sortear
botao_sortear = tk.Button(tela_principal, text="Iniciar Sorteio", command=sortear, width=30, height=2)
botao_sortear.place(x=300,y=365)

# Botão para exibir resultados registrados
botao_exibir_registrados = tk.Button(tela_principal, text="Exibir Resultados Registrados", command=exibir_resultados_registrados, width=30)
botao_exibir_registrados.place(x=300, y=410, height=50)

# Manter a tela em execução
tela_principal.mainloop()
```
